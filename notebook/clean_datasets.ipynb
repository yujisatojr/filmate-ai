```python
import pandas as pd

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('../datasets/sentiment_latest.csv')

# Drop the 'synopsis' column if it exists
if 'synopsis' in df.columns:
    df.drop(columns=['synopsis'], inplace=True)

df.head()
```

```python
if 'certificate' in df.columns:
    # Count the number of rows for each unique value in the 'certificate' column
    certificate_counts = df['certificate'].value_counts()
    
    # Print out unique values and their corresponding row counts
    print("Unique values in the 'certificate' column and their counts:")
    for certificate, count in certificate_counts.items():
        print(f"{certificate}: {count} row(s)")
else:
    print("Certificate column not found in the DataFrame.")
```

```python
from tqdm import tqdm
tqdm.pandas()

def combine_metadata(row):
    cast = ', '.join(filter(None, [str(row['cast_1']), str(row['cast_2']), str(row['cast_3'])]))
    return f"{row['title']} | {row['director_1']} | {cast} | {row['summary']}"

# Create new column 'metadata'
df['metadata'] = df.progress_apply(combine_metadata, axis=1)
df.head()
```

```python
file_path = 'sentiment_latest_cleaned.csv'
df.to_csv(file_path, index=False)
```
