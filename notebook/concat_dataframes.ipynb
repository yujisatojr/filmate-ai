```python
import pandas as pd

# Read the dataframes
file_path = '../datasets/movies_final_wo_vector.csv'
df = pd.read_csv(file_path)
df = df.fillna('')

df_genres_casts = pd.read_csv('../datasets/genres_casts_latest.csv')
df_synopsis = pd.read_csv('../datasets/synopsis_latest.csv')

# Merge df_genres_casts into df
merged_df = pd.merge(df, df_genres_casts, on='id', how='left')

# Merge df_synopsis into the merged_df
merged_df = pd.merge(merged_df, df_synopsis, on='id', how='left')

# Output the length and first few rows of the merged dataframe
print(str(len(merged_df)) + ' rows')
merged_df.head()
```

```python
merged_df = merged_df.drop(columns=['genre'])
merged_df = merged_df.drop(columns=['sentiment_compound'])
merged_df = merged_df.drop(columns=['sentiment_normalized'])

other_columns = [col for col in merged_df.columns if col != 'img']

# Reorder columns so 'img' is at the very end
new_order = other_columns + ['img']

# Reassign DataFrame with the new column order
merged_df = merged_df[new_order]

merged_df.head()
```

```python
file_path = 'dataset.csv'
merged_df.to_csv(file_path, index=False)
```

```python
removed_df = merged_df.drop(columns=['synopsis'])
removed_df.to_csv('removed.csv', index=False)
```
