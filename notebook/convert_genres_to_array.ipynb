```python
import pandas as pd

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('../datasets/movies_embedding_051424.csv')
```

```python
# Function to extract genres from multiple columns and return as a list
def extract_genres(row):
    genres = []
    for col in ['genre_1', 'genre_2', 'genre_3']:
        genre = row[col]
        if pd.notnull(genre):
            genres.append(genre)
    return genres

# Apply the function to create a new 'genres' column
df['genres'] = df.apply(extract_genres, axis=1)
df.head()
```

```python
def extract_directors(row):
    directors = []
    for col in ['director_1', 'director_2', 'director_3']:
        director = row[col]
        if pd.notnull(director):
            directors.append(director)
    return directors

df['directors'] = df.apply(extract_directors, axis=1)
df.head()
```

```python
def extract_writers(row):
    writers = []
    for col in ['writer_1', 'writer_2', 'writer_3']:
        writer = row[col]
        if pd.notnull(writer):
            writers.append(writer)
    return writers

df['writers'] = df.apply(extract_writers, axis=1)
df.head()
```

```python
def extract_casts(row):
    casts = []
    for col in ['cast_1', 'cast_2', 'cast_3']:
        cast = row[col]
        if pd.notnull(cast):
            casts.append(cast)
    return casts

df['casts'] = df.apply(extract_casts, axis=1)
df.head()
```

```python
other_columns = [col for col in df.columns if col != 'img']

new_order = other_columns + ['img']

df = df.reindex(columns=new_order)
df.head()
```

```python
other_columns = [col for col in df.columns if col != 'metadata']

new_order = other_columns + ['metadata']

df = df.reindex(columns=new_order)
df.head()
```

```python
other_columns = [col for col in df.columns if col != 'metadata_vector']

new_order = other_columns + ['metadata_vector']

df = df.reindex(columns=new_order)
df.head()
```

```python
file_path = "movies_all_051424.csv"

df.to_csv(file_path, index=False)
```
