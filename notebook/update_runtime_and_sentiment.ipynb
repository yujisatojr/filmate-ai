```python
import pandas as pd

file_path = '../datasets/movies_vector.csv'
df = pd.read_csv(file_path)

df.head()
```

```python
def runtime_to_minutes(time_str):
    if 'h' in time_str:
        # Splitting hours and minutes
        parts = time_str.split('h ')
        # Extracting numerical part for hours
        hours = int(parts[0].replace('h', ''))
        if len(parts) > 1 and 'm' in parts[1]:
            minutes = int(parts[1][:-1])
        else:
            minutes = 0
    else:
        # If there are no hours, set it to 0
        hours = 0
        if 'm' in time_str:
            minutes = int(time_str[:-1])
        else:
            minutes = 0

    # Calculating total minutes
    total_minutes = hours * 60 + minutes
    return total_minutes

# Apply the function to each value in the 'runtime' column and create a new column 'runtime_mins'
df['runtime_mins'] = df['runtime'].apply(runtime_to_minutes)

# Print the DataFrame to verify the new column
print(df.head())
```

```python
df.head()
```

```python
unique_runtimes = df['runtime_mins'].unique()
for runtime in unique_runtimes:
    print(runtime)
```

```python
# Move 'runtime_mins' column after 'runtime'
runtime_mins_col = df.pop('runtime_mins')
df.insert(df.columns.get_loc('runtime') + 1, 'runtime_mins', runtime_mins_col)

df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Define the scaler
scaler = MinMaxScaler(feature_range=(1, 5))

# Fit and transform the data
scaled_sentiment = scaler.fit_transform(df[['sentiment']])

# Round the scaled values to the nearest integer
rounded_sentiment = [round(value) for value in scaled_sentiment.flatten()]

# Store the rounded values in a new column
df['sentiment_normalized'] = rounded_sentiment
df.head()
```

```python
normalized_sentiment = df.pop('sentiment_normalized')

# Insert the 'normalized_sentiment' column after the 'sentiment' column
df.insert(df.columns.get_loc('sentiment') + 1, 'sentiment_normalized', normalized_sentiment)

# Print the DataFrame with reordered columns
df.head()
```

```python
df.rename(columns={'sentiment': 'sentiment_compound'}, inplace=True)
df.head()
```

```python
df.to_csv("movies_final.csv", index=False)
```
